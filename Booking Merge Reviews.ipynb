```python
import os
import pandas as pd
```

```python
# Path to the folder containing Excel files
folder_path = r'G:\Booking Scraping\Reviews\Positive Reviews'

# List to store individual DataFrames
dfs = []

# Loop through each file in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.xlsx') or file_name.endswith('.xls'):
        file_path = os.path.join(folder_path, file_name)
        # Read the Excel file and append the DataFrame to the list
        df = pd.read_excel(file_path)
        dfs.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
merged_df = pd.concat(dfs, ignore_index=True)

# Optional: Save the merged DataFrame to a new Excel file
output_file_path = r'G:\Booking Scraping\Reviews\Positive Reviews\merged_reviews.xlsx'
merged_df.to_excel(output_file_path, index=False)

print("All Excel files merged successfully.")
```
