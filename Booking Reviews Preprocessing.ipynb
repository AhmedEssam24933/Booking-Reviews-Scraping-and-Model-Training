```python
import pandas as pd
import re
```

```python
negative_reviews = pd.read_excel(r'J:\Booking Scraping\Reviews\Negative Reviews\merged_reviews.xlsx')
negative_reviews
```

```python
# Function to remove English words
def remove_english(text):
    # Ensure the input is a string
    text = str(text)
    # Regular expression pattern to match English words
    pattern = re.compile(r'[A-Za-z]+')
    # Substitute English words with an empty string
    return re.sub(pattern, '', text)

# Apply the function to the column
negative_reviews['Reviews'] = negative_reviews['Reviews'].apply(remove_english)
```

```python
# Function to remove non-Arabic characters while keeping spaces and replacing hyphens with spaces
def remove_non_arabic(text):
    if pd.isna(text):  # Check for NaN values
        return text
    # Replace hyphens and newlines with spaces
    text = text.replace('-', ' ').replace('\n', ' ')
    # Regular expression pattern to match non-Arabic characters (except spaces)
    pattern = re.compile(r'[^\u0600-\u06FF\u0750-\u077F\u08A0-\u08FF\s]+')
    # Substitute non-Arabic characters with an empty string
    return re.sub(pattern, '', text)

# Apply the function to the column
negative_reviews['Reviews'] = negative_reviews['Reviews'].apply(remove_non_arabic)
```

```python
# Remove rows that are entirely empty or contain only whitespace
negative_reviews = negative_reviews[negative_reviews['Reviews'].str.strip() != '']
```

```python
phrases_to_remove = [
    "لا يوجد", "لا شي", "لا شيء", "لاشيء", "لاشي", "ولاشي", "ولا شيء", "ولا شي",
    "لايوجد", "كل شي", "كل شيء", "لاشئ", "ولاشئ", "لا شئ", "ولا شئ"
]


# Remove rows containing the specific phrases
negative_reviews = negative_reviews[~negative_reviews['Reviews'].isin(phrases_to_remove)]
negative_reviews = negative_reviews.reset_index(drop = True)
negative_reviews
```
